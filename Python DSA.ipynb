# Python Data Structures

This notebook provides a professional overview of core Python data structures including Lists, Tuples, Dictionaries, and Sets.

## 1. Lists 
- Ordered
- mutable collections
- allow duplicates

```python
my_list = [1, 2, 3, 3, 'AI', 5.5]
my_list.append(10)
my_list.remove(2)
print('List:', my_list)
print('First element:', my_list[0])
```

## 2. Tuples 
- Ordered
- immutable collections
- allow duplicates

```python
my_tuple = (1, 2, 3,3)
print('Tuple:', my_tuple)
print('Second element:', my_tuple[1])
```

## 3. Dictionaries 
Key-value pair collections with fast lookup.

```python
my_dict = {
    'name': 'Kurup',
    'age': 23,
    'role': 'AI Engineer'
}
print('Name:', my_dict['name'])
my_dict['age'] = 24
print('Updated Dictionary:', my_dict)
```

## 4. Sets 
- Unordered collections of unique elements.

```python
my_set = {1, 2, 3, 3}
print('Set (duplicates removed):', my_set)
my_set.add(4)
print('Updated Set:', my_set)
```

## Comparison Summary

| Feature | List | Tuple | Dictionary | Set |
|--------|------|------|-----------|-----|
| Order | Yes | Yes | Yes (3.7+) | No |
| Mutable | Yes | No | Yes | Yes |
| Duplicates | Yes | Yes | Keys No | No |
| Indexing | Yes | Yes | No | No |
| Speed | Medium | Fast | Very Fast | Fast |
