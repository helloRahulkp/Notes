# 🧠 Autograd Demo

This notebook demonstrates the core functionality of the Value class.

## Topics
- Basic operations
- Computational graph
- Backpropagation

```python
from micrograd.engine import Value
```

## ➕ Basic Operations

```python
a = Value(2.0)
b = Value(3.0)

c = a + b
d = a * b

print('c:', c)
print('d:', d)
```

## 🔄 Backward Pass

```python
x = Value(2.0)
y = Value(3.0)

z = x * y + x
z.backward()

print('z:', z)
print('dz/dx:', x.grad)
print('dz/dy:', y.grad)
```

## ⚡ Activation Function

```python
a = Value(0.5)
t = a.tanh()
t.backward()

print('tanh:', t)
print('grad:', a.grad)
```
