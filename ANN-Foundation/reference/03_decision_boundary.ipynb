# 🌙 Decision Boundary Visualization

Train an MLP on a non-linear dataset and visualize decision boundary.

## Concepts
- Non-linear classification
- Model generalization
- Visualization

```python
from micrograd.nn import MLP
from micrograd.utils import make_moons_dataset, plot_decision_boundary
```

```python
def make_moons_dataset(n_samples=200, noise=0.1):
    return make_moons(n_samples=n_samples, noise=noise)
```

```python
def plot_decision_boundary(model, X, y):
    h = 0.05
    x_min, x_max = X[:, 0].min()-1, X[:, 0].max()+1
    y_min, y_max = X[:, 1].min()-1, X[:, 1].max()+1

    xx, yy = np.meshgrid(
        np.arange(x_min, x_max, h),
        np.arange(y_min, y_max, h)
    )

    grid = np.c_[xx.ravel(), yy.ravel()]
    preds = [model(list(p)).data for p in grid]
    Z = np.array(preds).reshape(xx.shape)

    plt.contourf(xx, yy, Z > 0, alpha=0.5)
    plt.scatter(X[:, 0], X[:, 1], c=y)
    plt.title("Decision Boundary")
    plt.show()
```

## 📊 Dataset

```python
X, y = make_moons_dataset(200, noise=0.2)
y = y * 2 - 1
```

## 🏗️ Model

```python
model = MLP(2, [16,16,1])
```

## 🔥 Training

```python
for epoch in range(50):
    total_loss = 0

    for xi, yi in zip(X, y):
        pred = model(list(xi))
        loss = (pred - yi)**2

        model.zero_grad()
        loss.backward()

        for p in model.parameters():
            p.data += -0.05 * p.grad

        total_loss += loss.data

    print(f"Epoch {epoch}: Loss {total_loss:.4f}")
```

## 📈 Visualization

```python
plot_decision_boundary(model, X, y)
```
